```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('paises.csv', delimiter=';')

NorthAm = df[df['Region'].str.contains('Northern America',case=False,na=False)]

paises = NorthAm['Country'].values
deathNA = NorthAm['Deathrate'].values
birthNA = NorthAm['Birthrate'].values

plt.figure(figsize=(12, 6))
plt.plot(paises, birthNA, marker='o', label='Taxa de Natalidade (Birthrate)')
plt.plot(paises, deathNA, marker='o', label='Taxa de Mortalidade (Deathrate)')
plt.xticks(rotation=45)
plt.xlabel('Países da América do Norte')
plt.ylabel('Taxas (%)')
plt.title('Taxa de Natalidade e Mortalidade - América do Norte')
plt.legend()
plt.show()

```

```python
df2 = pd.read_csv('space.csv', delimiter=';')

usa = df2[df2['Location'].str.contains('usa',case=False,na=False)]
china = df2[df2['Location'].str.contains('china',case=False,na=False)]

plt.title('Quantidade de empresas diferentes nos USA e na China')
plt.bar(['USA','China'],[usa['Company Name'].nunique(), china['Company Name'].nunique()], color=['blue', 'red'])
```

```python
roscosmos = df2[df2['Company Name'].str.contains('roscosmos',case=False,na=False)]
roscSuccess = roscosmos[roscosmos['Status Mission'] == 'Success']
roscFeilure = roscosmos[roscosmos['Status Mission'] == 'Failure']
sucesso = roscSuccess['Num'].size
falha = roscFeilure['Num'].size
plt.title('Missões feitas pela Companhia Roscosmos')
plt.pie(x = [sucesso,falha], labels=['% De sucesso ','% De falhas'],autopct='%1.1f%%')
```
